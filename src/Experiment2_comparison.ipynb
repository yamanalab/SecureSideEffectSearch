```python
import pandas as pd

data1 = pd.read_csv( 'exp3_1.tsv', delimiter='\t' )
data7 = pd.read_csv( 'exp3_7.tsv', delimiter='\t' )
data14 = pd.read_csv( 'exp3_14.tsv', delimiter='\t' )
data28 = pd.read_csv( 'exp3_28.tsv', delimiter='\t' )

import numpy as np

import pylab

#data = data2[data2['percentage']<10.]
#data = data2
data1['percentage'] *= 400
data7['percentage'] *= 400
data14['percentage'] *= 400
data28['percentage'] *= 400

#print(len(data2)/len(data))

pylab.figure(figsize=(9,6))
pylab.xlim([0,8000])
pylab.ylim([0,400])
pylab.xlabel(u'Filtered Database Size (PCS)', fontsize=15)
pylab.ylabel(u'Time (sec)', fontsize=15)
pylab.title(u'Time Spent Per Query (Randomly Generated)', fontsize=15)

pylab.scatter(x=data1['percentage'], y=data1['excecution_time'], marker = 'o', label=u'No multithreading')
pylab.scatter(x=data7['percentage'], y=data7['excecution_time'], marker = '^', label=u'Maximum 7 Thread')
pylab.scatter(x=data14['percentage'], y=data14['excecution_time'], marker = '*', label=u'Maximum 14 Thread')
pylab.scatter(x=data28['percentage'], y=data28['excecution_time'], marker = 'H', label=u'Maximum 28 Thread')
pylab.legend(fontsize=15)
```
